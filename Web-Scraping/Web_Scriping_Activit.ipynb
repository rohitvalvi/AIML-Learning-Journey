```python
# Import required libraries

import requests
from bs4 import BeautifulSoup

```

```python

page_count = 1

while True:
    # URL of the website to scrape
    URL = f"https://quotes.toscrape.com/page/{page_count}/"
    
# Sending an HTTP GET request to the specified URL
    res = requests.get(URL)

# Parsing the HTML content of the webpage using BeautifulSoup
    soup = BeautifulSoup(res.text, "lxml")
    quotes = soup.select("div.quote")
    
    if not quotes:
        print("No Valid pages")
        break

    with open(f"Quotes_data/data{page_count}.html", "w") as f:
        f.write(res.text)
        print("Downloaded")
    page_count = page_count + 1
```

```python

with open("Quotes_data/data.html", "r") as f:
    html_content = f.read()

soup = BeautifulSoup(html_content, "lxml")
```

```python
all_quotes = soup.select("div.quote")
life_quotes = []

for q in all_quotes:
    all_tags = []

    for tag in q.select(".tags .tag"):
        all_tags.append(tag.get_text())

    if "life" in all_tags:
        text = q.select_one("span.text").get_text()
        author = q.select_one("small.author").get_text()
        life_quotes.append([text, author])

life_quotes
```

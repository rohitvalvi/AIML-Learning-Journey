```python
# import Libraries
import pandas as pd
import seaborn as sns
```

```python
# Reading data from file
insurance = pd.read_csv("insurance.csv")
insurance
```

```python
# Plot data for better Visualization
sns.scatterplot(data = insurance, x="bmi", y="charges", hue="smoker")
# Graph show us that the snoker people have higher charges compare to Non Smoker people
```

```python
sns.scatterplot(data=insurance, x="age", y="charges" , hue="smoker")
```

```python
# Spliting data for Train and prediction data
X = insurance.drop(columns = ["charges", "region"])
Y = insurance["charges"]

# Convert string data to Numerical data
X["sex"] = X["sex"].map({"female": 1, "male": 0})
X["smoker"] = X["smoker"].map({"yes": 1, "no": 0})
```

```python
X.head()
```

```python
# Split data for t=Train and Test(Train = 80% , Test = 20%)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42)
```

```python
# Train the Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predicting the Values 
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Test the Data
from sklearn.metrics import r2_score

r2_score(y_test,y_pred )
```
